```python
# Imports
import matplotlib.pyplot as plt
import maze_dataset as md
import maze_generator.maze_dataset as md
import numpy as np
import DQN.training as training
import DQN.agent as agent
import DQN.models as models
import DQN.hyperparameter_fixing as hyper
import Maze_env.wrappers.rewards as rw
import os
print(os.getcwd())
```

```python
dataset = md.Maze_dataset(1,(3,3),maze_type = 'percolation')
```

```python
rewards = rw.reward_dist(GOAL = 20.0,
                         SEE_GOAL = 0.01,
                         DONT_SEE_GOAL = -0.005,
                         NEW_PLACE = 0.01,
                         OLD_PLACE = -0.005,
                         GET_CLOSER = 0.05,
                         GET_FARTHER = -0.005,
                         DIST = 0.0)

hp = hyper.agent_tuning(name='hp_test_rw',n_iter=5,
                        maze_dataset=dataset,
                        model=models.CNN_version1,
                        vision=3,
                        reward=rewards,
                        len_game=30,
                              gamma = 0.99,
                              tau = 0.001,
                               batch_size = 64,
                              n_frames = 500000,
                              lr = 0.0005,
                              lr_step_size=10000,
                              lr_gamma = 0.1,
                              replay_buffer_min_perc=0.1,
                              lambda_entropy=0.4,
                              beta = 0.4,
                              alpha = 0.6,
                              curriculum=True
                        )
```

```python
load = agent.maze_agents.load('hp_test2')
testing_dataset = md.Maze_dataset(1,shape=(3,3),maze_type='percolation')
load.run_agent(testing_dataset[0],len_game=15,n_episodes=20,agents_pos = None, targets_pos =None)
```
