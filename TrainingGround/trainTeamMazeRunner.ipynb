## Training area for agents in Team Maze Runner

```python
# Imports
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
import matplotlib.pyplot as plt
import maze_generator.maze_dataset as md
import numpy as np
from DQN.training.basic import BaseTraining
from DQN.agents.basic import BaseAgent
from Maze_env.game_info.team_maze_runner_info import TeamMazeRunner

import DQN.models.base as base
print(os.getcwd())
```

### Define the Mazes

```python
dataset = md.Maze_dataset(1,(3,3),maze_type = 'percolation')
```

```python
# example of the maze
dataset.show_maze(0)
```

```python
# parameters for the agents
n_agents = 1
vision = 3
```

```python
%matplotlib notebook


# --- Agents for the Hunger Games --- #
maze_agent =BaseAgent({'team_0':base.CNN_version1,
                       'team_1':base.CNN_version1},
                               vision={'team_0':vision,'team_1':vision},
                               action_type='cardinal',
                               dist_paradigm='path',
                               game_info=TeamMazeRunner(n_teams=2)
                               )
maze_agent.load_object('team_1','MazeRunner','agents')

# --- trainer of the agent --- #
train = BaseTraining(name = 'TeamMazeRunner',
                              maze_dataset = dataset,
                              maze_agent = maze_agent,
                              len_game=30,
                              n_objects={'team_0':n_agents,'team_1':n_agents},
                              final_epsilon = 0.05,
                              gamma = 0.99,
                              tau = 0.0001,
                               batch_size = 64,
                              lr = 0.0005,
                              lr_step_size=5000,
                              lr_gamma = 0.98,
                              l2_regular=0.01,
                              replay_buffer_size=400000,
                              replay_buffer_min_perc=0.1,
                              target_update=5000,
                              policy_update=1,
                              lambda_entropy=0.1,
                              beta = 0.4,
                              alpha = 0.6,
                              decay_total = 400000,
                              per = True,
                              frame_mult=1.5,
                              )
train.training_toggle_off(['team_1'])
train.train(test_agent=False,peak=False)
```

```python
train.save()
train.results()
```

```python
train.agents.test_agent(dataset,n_episodes=1000,len_game=15,num_objects={'team_0':n_agents,'team_1':n_agents})
```

```python

train.agents.run_agent(dataset[0],num_objects={'team_0':n_agents, 'team_1':n_agents},n_episodes=20,len_game=15,epsilon=0,init_pos={})
```
